# 2-D GRAPH


**HOW TO MADE 2-D GRAPH**

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([3,4])
y=np.array([2,7])
plt.plot(x,y,marker='o')
plt.show()
```

```python
#HOW TO FIND THE DISTANCE FROM ORIGAIN
def distance(x1,x2):
     return np.sqrt((x1)**2 +(x2)**2)
print(distance(3,4))  # Example usage
```

```python
# how to find the difference between two point
def distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
print(distance(3, 4, 2, 7))  # Example usage
```
