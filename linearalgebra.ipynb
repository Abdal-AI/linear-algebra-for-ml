```python
# how to find the difference between two point
def distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
print(distance(3, 4, 2, 7))  # Example usage
```

# HOW TO FIND ANGLE BETWEEN TO POINT OR VECTOR


```python
import numpy as np 
import matplotlib.pyplot as plt
x=np.array([1,2])
y=np.array([3,5])
dot_product=np.dot(x,y)
print("the dot product is :",dot_product)
# now find mignitude
mignitude_x=np.linalg.norm(x)
print("mignitude of x is :",mignitude_x)
mignitude_y=np.linalg.norm(y)
print("mignitude of x is :",mignitude_y)
# then find the angle
angle_radian=np.arccos(dot_product/(mignitude_x*mignitude_y))
print("angle in radian is :",angle_radian)
degree=np.degrees(angle_radian)
print(f"Angle between A and B: {degree:.2f} degrees")


```
